```python
import pandas as pd 
import plotly.express as px
import numpy as np
```

```python
df = pd.read_csv("games.csv")
print(df.head())
df.info()
```

```python
df.columns = df.columns.str.lower()
print(df.head())
```

<h2 style="color: teal;">Changed all column names to lowercase.</h2>

```python
df.info()
```

<h2 style="color: teal;">Column dtype Changes:</h2> <br>
1. na_sales to int64 after rounding  <br>
3. eu_sales to int64 after rounding  <br>
4. jp_sales to int64 after rounding  <br>
5. other_sales to int64 after rounding<br>
*All Sales data converted to int64 for mostly graphing purposes* <br>
Following columns have missing values: <br>
6. critic_score to int64 after removing .0  <br>
7. user_score to int64 after potentially rounding (only goes to 10, so could potentially keep the decimal if it doesn't mess up graphing)  <br>
8. rating (no column dtype change, but has missing values)  <br>
9. year_of_release just needs to drop .0; no real need to change to datetime or int64.

```python
df['na_sales'] = df['na_sales'].round().astype('int64')
df['eu_sales'] = df['eu_sales'].round().astype('int64')
df['jp_sales'] = df['jp_sales'].round().astype('int64')
df['other_sales'] = df['other_sales'].round().astype('int64')
df['user_score'] = df['user_score'].round()


display(df.info())
display(df.head())
```

```python
import plotly.express as px
fig = px.scatter(df, x=df.index, y='year_of_release', title='Scatter Plot for Outlier Detection')
fig.show()
fig = px.box(df, y='year_of_release', title='Box Plot for Outlier Detection')
fig.show()
```

```python
year_mean = df['year_of_release'].mean()
print(year_mean)
```

```python
median_value = df['year_of_release'].median()
df['year_of_release'] = df['year_of_release'].fillna(median_value)
df['year_of_release'] = df['year_of_release'].astype('int64')
df.info()
```

<h2 style="color: teal;">Filled missing values with median, since a good amount of outliers compared to the overall data of this column. But mean is only a year off so it doesn't matter all that much. However, using the mean does make it easier on me since the mean is a decimal number. </h2>

```python
fig = px.scatter(df, x=df.index, y='user_score', title='Scatter Plot for User_score Outlier Detection')
fig.show()
fig = px.box(df, y='user_score', title='Box Plot for user_score Outlier Detection')
fig.show()
fig = px.scatter(df, x=df.index, y='critic_score', title='Scatter Plot for critic_score Outlier Detection')
fig.show()
fig = px.box(df, y='critic_score', title='Box Plot for critic_score Outlier Detection')
fig.show()
```

```python
critic_mean = df['critic_score'].mean()
print(critic_mean)
```

```python
critic_median = df['critic_score'].mean()
df['critic_score'] = df['critic_score'].fillna(critic_median)
df['critic_score'] = df['critic_score'].astype('int64')
df.info()
```

<h2 style="color: teal;">Same as year_of_release. Median and mean are very close. Median does not have a decimal. </h2>

```python
tbd_count = (df['user_score'] == 'tbd').sum()
null_count = df['user_score'].isna().sum()
total_missing = tbd_count + null_count
print("TBD Count:", tbd_count)
print("Null Count:", null_count)
print("Total Missing Values:", total_missing)
print()
df.info()
```

```python
df['user_score'] = df['user_score'].replace('tbd', np.nan)

df['user_score'] = pd.to_numeric(df['user_score'], errors='coerce')

median_value = df['user_score'].median()
df['user_score'] = df['user_score'].fillna(median_value)

print(df['user_score'])
print(df.info())
```

<h2 style="color: teal;">Changed all tbd values to NaN values, then changed all NaN to median. </h2>

```python
df['total_sales'] = df[['na_sales', 'eu_sales', 'jp_sales', 'other_sales']].sum(axis=1)

print(df[['na_sales', 'eu_sales', 'jp_sales', 'other_sales', 'total_sales']])
```

```python
duplicates = df[df.duplicated()]
display(f"Number of duplicate rows: {len(duplicates)}")
display(duplicates)
```

<h2 style="color: teal;">Missing data most likely because of human error, or unavailable data as in the case of all the 'tbd' values. </h2>

```python
#grouping by release year
games_per_year = df.groupby('year_of_release')['name'].count()
print(games_per_year)
```

```python
games_per_year_df = games_per_year.reset_index()
games_per_year_df.columns = ['year', 'count']
display(games_per_year_df)

fig = px.bar(games_per_year_df, x='year', y='count', 
             title='Number of Games Released Per Year',
             labels={'year': 'Year', 'count': 'Number of Games'})
fig.show()
```

```python
#when did each platfrom launch
platform_launch = df.groupby('platform')['year_of_release'].min()
platform_launch_df = platform_launch.reset_index()
platform_launch_df.columns = ['platform', 'year']
display(platform_launch_df)
```

```python

# Combining Data

combined_df = pd.merge(
    games_per_year_df, 
    platform_launch_df, 
    on='year', 
    how='outer')
combined_df.fillna('None Released', inplace=True) 
display(combined_df)
```

<h2 style="color: teal;">I wanted to see if the jumps in games released year-over-year were correlated with the release of a new platform. It seems like it might have something to it.</h2>

```python
platform_sales = df.groupby('platform')['total_sales'].sum().reset_index()

platform_sales = platform_sales.sort_values(by='total_sales', ascending=False)

fig = px.bar(
    platform_sales,
    x='platform',
    y='total_sales',
    title='Total Sales by Platform',
    labels={'platform': 'Platform', 'total_sales': 'Total Sales'},
    text='total_sales')

fig.update_layout(xaxis_tickangle=45, template='plotly_white')

fig.show()
```

```python

top_platforms = platform_sales.head(10)['platform'].tolist() 

top_platforms_data = df[df['platform'].isin(top_platforms)]

sales_by_year_platform = top_platforms_data.groupby(['year_of_release', 'platform'])['total_sales'].sum().reset_index()


fig = px.line(
    sales_by_year_platform,
    x='year_of_release',
    y='total_sales',
    color='platform',
    title='Yearly Sales Distribution for Top Platforms',
    labels={'year_of_release': 'Year of Release', 'total_sales': 'Total Sales','platform': 'Platform'})


fig.update_layout(xaxis_title='Year', yaxis_title='Total Sales', legend_title='Platform', template='plotly_white')

fig.show()
```

<h2 style="color: teal;">The top five platforms that created the biggest revenue for game sales, all now have zero sales.</h2>

```python
platform_end_df = df.groupby('platform')['year_of_release'].max().reset_index()
platform_end_df.rename(columns={'year_of_release': 'end_year'}, inplace=True)

platform_lifespan_df = pd.merge(platform_launch_df, platform_end_df, on='platform')

platform_lifespan_df['lifespan'] = platform_lifespan_df['end_year'] - platform_lifespan_df['year'] + 1

print(platform_lifespan_df)
```

```python

average_lifespan = platform_lifespan_df['lifespan'].mean()

print(f"Average platform lifespan: {average_lifespan:.2f} years")
```

```python
platform_sorted = platform_lifespan_df.sort_values(by='end_year', ascending=False)
platform_sorted['total_sales'] = df.groupby('platform')['total_sales'].sum().reset_index(drop=True)

display(platform_sorted)
```

<h2 style="color: teal;">Since the platforms that had the highest sales don't bring in any revenue for the company anymore, I need to look at platforms that are still active in 2016, and will hopefully be active in 2017. Going to focus on trends from X360, PC, XOne, PS4, and PS3, and most likely choose the previous five years as my period to predict total sales for 2017. </h2>

```python

filtered_data = df[(df['year_of_release'] >= 2012) & (df['year_of_release'] <= 2016)]

display(filtered_data.head(20))
```

```python
total_sales_by_platform = filtered_data.groupby('platform')['total_sales'].sum().reset_index()

total_sales_by_platform = total_sales_by_platform.sort_values(by='total_sales', ascending=False)

fig1 = px.bar(total_sales_by_platform, 
              x='platform', 
              y='total_sales', 
              title='Total Sales by Platform', 
              labels={'platform': 'Platform', 'total_sales': 'Total Sales'},
              color='total_sales',
              color_continuous_scale='Viridis')

fig1.update_layout(xaxis_title="Platform", yaxis_title="Total Sales")
fig1.show()
```

<h2 style="color: teal;">The top five platforms in the last five years are PS4, X360, PS3, 3DS, and XOne.</h2>

```python
sales_by_platform_year = filtered_data.groupby(['platform', 'year_of_release'])['total_sales'].sum().reset_index()


fig2 = px.line(sales_by_platform_year, 
               x='year_of_release', 
               y='total_sales', 
               color='platform', 
               title='Sales Trend by Platform Over Time', 
               labels={'year_of_release': 'Year', 'total_sales': 'Sales'},
               line_shape='linear')

fig2.update_layout(xaxis_title="Year", yaxis_title="Total Sales")
fig2.show()
```

```python
na_sales = filtered_data.groupby('platform')['na_sales'].sum().reset_index()
eu_sales = filtered_data.groupby('platform')['eu_sales'].sum().reset_index()
jp_sales = filtered_data.groupby('platform')['jp_sales'].sum().reset_index()

na_sales['region'] = 'NA'
eu_sales['region'] = 'EU'
jp_sales['region'] = 'JP'

regional_sales = pd.concat([na_sales, eu_sales, jp_sales], ignore_index=True)
display(regional_sales)
```

```python
top_na_platforms = (regional_sales.sort_values(['region', 'na_sales'], ascending=[True, False]).groupby('region').head(5))


display(top_na_platforms)
```
